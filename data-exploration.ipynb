```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

DATA_DIR = "D:/Machine-Learning/HousePrice prediction/data"
def read_data():
    return pd.read_csv(f"{DATA_DIR}/Housing.csv")
```

```python
def price_distribution(df):
    plt.figure(figsize=(8,5))
    sns.histplot(df['price'], kde=True)
    plt.title('Distribution of House Prices')
    plt.show()
```

```python
def area_distribution(df):
    plt.figure(figsize=(8,5))
    sns.histplot(df['area'], kde=True)
    plt.title('Distribution of House Areas')
    plt.show()
```

```python
def price_area_relation(df):
    print(df[['area', 'price']].isnull().sum())
    plt.figure(figsize=(8,5))
    sns.regplot(x='area', y='price', data=df, scatter_kws={'alpha':0.6})

    plt.title('Price vs Area with Trend Line')
    plt.show()
```

```python
def price_bedrooms_relation(df):
    print(df[['bedrooms', 'price']].isnull().sum())
    plt.figure(figsize=(8,5))
    sns.boxplot(x='bedrooms', y='price', data=df)

    plt.title('Price vs Number of Bedrooms')
    plt.show()
```

```python
def price_stories_relation(df):
    print(df[['stories', 'price']].isnull().sum())
    plt.figure(figsize=(8,5))
    sns.boxplot(x='stories', y='price', data=df)

    plt.title('Price vs Number of Stories')
    plt.show()
```

```python
def correlation_heatmap(df):
    plt.figure(figsize=(10,8))
    sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
    plt.title('Correlation Heatmap')
    plt.show()
```

```python
def price_vs_categorical(df, col):
    plt.figure(figsize=(8,5))
    sns.boxplot(x=col, y='price', data=df)
    plt.title(f'Price vs {col}')
    plt.show()
```

```python
def outliers_analysis(df, col):
    plt.figure(figsize=(8,5))
    sns.boxplot(x=df[col])
    plt.title(f'Outliers in {col}')
    plt.show()
```

```python

df = read_data()
print(df.head(10))
print(df.describe())
```

```python
price_distribution(df)
```

```python
area_distribution(df)
```

```python
price_area_relation(df)
```

```python
price_bedrooms_relation(df)
```

```python
price_stories_relation(df)
```

```python
correlation_heatmap(df)
```

```python
cols = ['airconditioning', 'prefarea', 'furnishingstatus']

for col in cols:
    price_vs_categorical(df, col)
```

```python
outliers_analysis(df, 'price')
outliers_analysis(df, 'area')
outliers_analysis(df, 'bedrooms')
outliers_analysis(df, 'stories')
outliers_analysis(df, 'parking')
outliers_analysis(df, 'bedrooms')
```
